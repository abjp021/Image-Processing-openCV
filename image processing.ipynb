## **Reading, Writing image files and performing image transformations such as rotation, scaling, filtering etc. using OpenCV.**

```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

# Reading and Displaying Image

```
img = cv2.imread("hello.png", 1) #colored
img = cv2.resize(img, (255, 255)) #Scaling the original Image to 255 x 255 px
cv2_imshow(img)

img1 = cv2.imread("hello.png", 0) #Gray Scale
img1 = cv2.resize(img1, (255, 255))
cv2_imshow(img1)

img2 = cv2.imread("hello.png", -1) #original
img2 = cv2.resize(img2, (255, 255))
cv2_imshow(img2)
```

#Writing Image

```
cv2.imwrite('copy_hello.png', img)
```

#Scaling 

```
original_img = cv2.imread("hello.png", -1) 
Scaled_img = cv2.resize(original_img, (255, 255)) #Scaling the original Image to 255 x 255 px
cv2_imshow(original_img)
cv2_imshow(Scaled_img)
cv2.imwrite('scaled_image.png',Scaled_img )
```

```
original_img = Scaled_img
cv2_imshow(original_img)
```

#Rotating Image

```
(rows, cols) = original_img.shape[:2] #getting shape of image
rot = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)  #getting matrix for rotation w.r.t center to 45 degree without scaling
rotation = cv2.warpAffine(original_img, rot, (cols, rows)) 
cv2_imshow(rotation)
cv2.imwrite('rotated_img.png',rotation)
```

## **Translating Image**
1.   Create translation matrix. 
2.   If the shift is (x, y) then matrix would be 
     M = [[1 0 x] , [0 1 y]]

3. If the shift is (x, y) then matrix would be 
     M = [[1 0 x] , [0 1 y]] 
### Let's shift by (100, 50). 

```
mat = np.float32([[1, 0, 100], [0, 1, 50]]) 
translated_img = cv2.warpAffine(original_img, mat, (cols,rows))
cv2_imshow(original_img)
cv2_imshow(translated_img)
cv2.imwrite("Translated_image.png", translated_img)
```

# Edge Detection

```
edge_img = cv2.Canny(original_img, 100,200)
cv2_imshow(original_img)
cv2_imshow(edge_img)
cv2.imwrite("Edge_image.png", edge_img)
```

# **Filtering**
## 1. Bilateral Filtering

```
original_img2 = cv2.imread("filter.jpeg", 1)
original_img2 = cv2.resize(original_img2, (255, 255)) 
kernel = np.ones((5,5),np.float32)/25  
bil_filter_image = cv2.bilateralFilter(original_img2,9,75,75)  
cv2_imshow(original_img2)
cv2_imshow(bil_filter_image)
cv2.imwrite("bilateral_filter_image.jpeg", bil_filter_image)
```

## 2. Filter2D

```
td_filter_img = cv2.filter2D(original_img2,-1,kernel)
cv2_imshow(original_img2)
cv2_imshow(td_filter_img)
cv2.imwrite("2D_filter_image.jpeg", td_filter_img)
```
